```python
#Importando bibliotecas
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Separação banco de dados: X e y
dados = { "Ano" : "Vendas",
        2010: 723,
        2011: 814,
        2012: 905,
        2014: 1087,
        2017: 1360,
        2018: 1451,
        2019: 1542
         }
X = list(dados.keys())
X.pop(0)
X_base = X
y = list(dados.values())
y.pop(0)
y_base = y
print(len(X_base),len(y_base))
```

```python
X_base = np.array(X_base).reshape(-1, 1)
modelo = LinearRegression()
modelo.fit(X_base, y_base)
# Coeficientes da regressão
coeficiente_angular = modelo.coef_[0]
coeficiente_linear = modelo.intercept_
```

```python
# Recebendo dados do usuário
# Criando o gráfico
plt.figure(figsize=(10, 5))
plt.plot(X_base, y_base, marker='o', linestyle='-', color='b')
plt.title('Quantidade de Vendas ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Vendas')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
x_teste = int(input("Insira o ano que deve ser previsto: "))
x_teste
X.append(x_teste)
x_teste = np.array(x_teste).reshape(-1, 1)
predicao = modelo.predict(x_teste)
predicao = predicao.tolist()
predicao = int(predicao[0])
y.append(predicao)
print(y)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(X, y, marker='o', linestyle='-', color='r')
plt.title('Quantidade de Vendas ao Longo dos Anos (Previsão)')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Vendas')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
